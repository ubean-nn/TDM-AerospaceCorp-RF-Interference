```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import os

# --- 1) collect all the .npy files
# (or build them manually if you know the range)
data_dir = "../fiot_highway2-main/data"
files = [f"{i:06d}.npy" for i in range(16479, 16479-100, -1)]  

# --- 2) load them and stack along axis=1 (x-direction) ---
arrays = [np.load(os.path.join(data_dir, f)) for f in files]

# check they all have the same number of rows!
for a in arrays[1:]:
    if a.shape[0] != arrays[0].shape[0]:
        raise ValueError("All arrays must have the same row count to stack horizontally.")

big_img = np.hstack(arrays)     # horizontal concatenation

# --- 3) plot ---
plt.figure(figsize=(12, 5))
plt.imshow(big_img, aspect='auto')  # aspect='auto' lets it stretch in x
plt.colorbar()
plt.show()
```
