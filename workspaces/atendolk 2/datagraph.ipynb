```python
import numpy as np
import os
import matplotlib.pyplot as plt

folder_path = "/anvil/projects/x-cis220051/corporate/aerospace-rf/fiot_highway2-main"
```

```python
selected_files = [
    "data/013287.npy", #0/a
    "data/015149.npy", #1/b
    "data/015548.npy", #2/c
    "data/000170.npy", #3/d
    "data/012864.npy", #4/e
    "data/006981.npy", #5/f
    "data/009074.npy", #6/g
    "data/002812.npy", #7/h
    "data/012947.npy"  #8/i
]

paths = [os.path.join(folder_path, f) for f in selected_files]

loaded_data = [(f, np.load(p)) for f, p in zip(selected_files, paths)]
print(f"Loaded {len(loaded_data)} files:")
for fname, data in loaded_data:
    print(f"  {fname}: shape={data.shape}, dtype={data.dtype}")
```

```python
import matplotlib.pyplot as plt

n = len(loaded_data)
fig, axes = plt.subplots(1, n, figsize=(4*n, 6), constrained_layout=True)

#print(plt.colormaps())

for ax, (fname, data) in zip(axes, loaded_data):
    vmin, vmax = np.percentile(data, 2), np.percentile(data, 98)
    im = ax.imshow(
        data,
        origin="lower",
        aspect="auto",
        vmin=vmin, vmax=vmax,
        cmap="gist_ncar"
    )
    ax.set_title(fname, fontsize=8)
    ax.set_xlabel("Time")
    ax.set_ylabel("Frequency")

plt.show()
```
